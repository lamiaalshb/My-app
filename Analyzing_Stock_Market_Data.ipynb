4. Are certain companies more volatile than others?

- Barplot

```python
# 4. Are certain companies more volatile than others?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_volatility(D1, D2):
    # Calculate the standard deviation (volatility) of the closing prices for each company
    volatility_1 = D1['Close'].std()  # Standard deviation for company 1 (Microsoft)
    volatility_2 = D2['Close'].std()  # Standard deviation for company 2 (Google)

    # Print the results
    print(f"Volatility of Company Microsoft: {volatility_1:.2f}")
    print(f"Volatility of Company Google : {volatility_2:.2f}")

    # Determine which company is more volatile
    if volatility_1 > volatility_2:
        print("Company Microsoft is more volatile.")
    elif volatility_1 < volatility_2:
        print("Company Google is more volatile.")
    else:
        print("Both companies have equal volatility.")

    # Create a bar plot for the volatility of each company
    companies = ['Microsoft', 'Google']
    volatilities = [volatility_1, volatility_2]

    plt.figure(figsize=(8, 5))
    sns.barplot(x=companies, y=volatilities, palette="Blues_d")
    plt.title("Volatility Comparison Between Microsoft and Google")
    plt.ylabel("Standard Deviation (Volatility)")
    plt.xlabel("Companies")
    plt.show()

# Call the function with the datasets D1 and D2 (Ensure D1 and D2 contain a 'Close' column)
calculate_volatility(D1, D2)

```

💡 The standard deviation for Microsoft (87.04) is higher than that of Google (12.62), indicating that Microsoft's stock prices experienced greater volatility and more deviation from the average compared to Google's. This means that Microsoft's stock prices showed larger fluctuations during the analyzed period, while Google's stock prices were less volatile and more stable.

-  Bollinger Bands

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your stock price data into a DataFrame (example: 'Close' column)
# Example: D1 is a DataFrame with stock data

# Calculate the moving average (middle band)
D1['MA20'] = D1['Close'].rolling(window=20).mean()

# Calculate the rolling standard deviation (for the bands)
D1['STD20'] = D1['Close'].rolling(window=20).std()

# Calculate the upper and lower Bollinger Bands
D1['Upper'] = D1['MA20'] + (D1['STD20'] * 2)
D1['Lower'] = D1['MA20'] - (D1['STD20'] * 2)

# Plot the closing price along with the Bollinger Bands
plt.figure(figsize=(12,6))
plt.plot(D1['Close'], label='Close Price', color='blue')
plt.plot(D1['MA20'], label='20-Day Moving Average', color='orange')
plt.plot(D1['Upper'], label='Upper Band', color='green', linestyle='--')
plt.plot(D1['Lower'], label='Lower Band', color='red', linestyle='--')

# Adding labels and title
plt.title('Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

💡 Benefit of Bollinger Bands Analysis in Decision Making**:

 **Technical Description of the Chart**:
- The **20-day Moving Average (MA20)** is the **yellow line** that reflects the general trend of the stock price over a 20-day period.
- The **Upper Band** and **Lower Band** represent the upper and lower boundaries of the stock price, calculated based on the **20-day standard deviation (STD20)**, which measures volatility. These bands serve as **reference points** showing the expected price levels under current market conditions.

**Volatility Analysis**:
1. **Narrow Bands**:
   - When the **Bollinger Bands** are narrow (close to the moving average), it indicates a period of **relative stability**, signaling **low volatility** in the market. In such cases, it could be a good time for investors seeking stability in their investments.
   
2. **Wide Bands**:
   - When the **Bollinger Bands** are wide (far from the moving average), it suggests **increased volatility**, signaling periods of **speculation or unstable market conditions**. This could be an opportunity for investors looking to **capitalize on market fluctuations** for short-term gains.
💡
- **When the Bollinger Bands are narrow**, this reflects relative **stability** in the **stock prices** of Microsoft and Google, which may be a signal for **conservative investors** seeking investments with low volatility.
- **When the Bollinger Bands are wide**, it could indicate heightened volatility, meaning there may be opportunities for **risk-taking investors** to benefit from **large price movements**. However, these investors should be prepared for market fluctuations.

Recommendation:
- **Conservative Investors**: Prefer periods of low volatility when the bands are narrow, as they can achieve stable returns without sudden market shifts.
- **Risk-Taking Investors**: Can make investment decisions based on the expansion of the bands, taking advantage of opportunities that arise during periods of high volatility.

 **In Conclusion**:
By using this analytical tool, investors can make informed decisions about **when to enter or exit** the market based on the **current level of volatility**.
